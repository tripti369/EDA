```python
# fill the null value with mode
df['student_id']= df['student_id'].fillna(df["student_id"].mode())
```

```python
for i in df.columns:
  print(f'{i}>>{df[i].mode()}')
```

```python
df['student_id']= df['student_id'].fillna(df["student_id"].mode()[0])
df['age']=df['age'].fillna(df["age"].mode()[0])
df['gender']=df['gender'].fillna(df["gender"].mode()[0])
df['grade']=df['grade'].fillna(df["grade"].mode()[0])
df['math_score']=df['math_score'].fillna(df["math_score"].mode()[0])
df['reading_score']=df['reading_score'].fillna(df["reading_score"].mode()[0])
df['writing_score']=df['writing_score'].fillna(df["writing_score"].mode()[0])
df['attendance_percentage']=df['attendance_percentage'].fillna(df["attendance_percentage"].mode()[0])
df['study_hours']=df['study_hours'].fillna(df["study_hours"].mode()[0])
df['sports_participation']=df['sports_participation'].fillna(df["sports_participation"].mean())
df['parent_income']=df['parent_income'].fillna(df["parent_income"].mode()[0])
df['school_type']=df['school_type'].fillna(df["school_type"].mode()[0])
df['city']=df['city'].fillna(df["city"].mode()[0])
```

```python
for i in df.columns:
  if df[i].dtypes==int or df[i].dtypes==float:
    df[i]=df[i].fillna(df[i].mean())
  else:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
# now check the datatypes

df['student_id']=df['student_id'].astype('int')
```

```python
df['math_score'] = pd.to_numeric(df['math_score'], errors='coerce')
df['math_score'] = df['math_score'].astype(float)
```

```python
df.dtypes
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'] = df['age'].astype('Int64') # Use Int64 to allow for potential NaN values after coercion
```

```python
df['grade'] = pd.to_numeric(df['grade'], errors='coerce')
df['grade'] = df['grade'].astype('Int64') # Use Int64 to allow for potential NaN values after coercion
```

```python
# removing outliers
per_25=df["math_score"].quantile(0.25)
per_75=df["math_score"].quantile(0.75)
iqr=per_75-per_25
lower_bound=per_25-(1.5*iqr)
upper_bound=per_75+(1.5*iqr)
df=df[(df["math_score"]>=lower_bound) & (df["math_score"]<=upper_bound)]
```

```python
# finding duplicate values
df['student_id'].duplicated().sum()
```

```python
# remove duplicate
df=df.drop_duplicates(subset='student_id')
```

```python
# finding duplicate values
df['student_id'].duplicated().sum()
```

```python
# sort the values basis
df=df.sort_values(by='math_score',ascending=False)
```

```python
# check the count of null then than 5%
null_count = df.isnull().sum()
null_percentage = (null_count / len(df)) * 100
print(null_percentage)
```

```python
# check if the count of null values are less than 5%
def check_null_values(df):
  for i in df.columns:
    null=df[i].isnull().sum()
    null_per= null*100/df.shape[0]
    if null_per<5:
      print(i,null_per)
      df[i] = df[i].dropna
```

```python
check_null_values(df)
```

```python
for col in df.columns:
  if df[col].isnull().sum()*100/len(df)<5:
    print(df[col].isnull().sum()*100/len(df))
    df.dropna(subset = [col],inplace=True)
```
